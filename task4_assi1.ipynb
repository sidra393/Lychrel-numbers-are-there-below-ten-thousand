```python
import math

def reverse(x):
  n = 0
  while x > 0:
    n = n * 10 + x % 10
    x = x // 10
  return n

def isPalindrome(n):
  return reverse(n) == n
```

```python
def isLychrel(n):
  for i in range(50):
    n+= reverse(n)
    if isPalindrome(n):
      return False
  return True
```

```python
cnt = 0

for i in range(10000):
  if isLychrel(i):
    cnt+= 1

print("Lychrel numbers Below 10000:",cnt)
```
